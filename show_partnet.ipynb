```python
import math
import pathlib

import h5py
import numpy as np
from plotly import graph_objects as go
import plotly.io as pio
from plotly.subplots import make_subplots
```

```python
pio.templates.default = "plotly_dark"
```

```python
%load_ext autoreload
%autoreload 2

from training.data.partnet import HdfIO
```

```python
num_samples = 9
```

```python
file = h5py.File("./datasets/raw/partnet/train.h5", "r") 
```

```python
file.keys()
```

```python
samples = file["pos"][:num_samples]
```

```python
seg_map = file["label_seg"][:num_samples]
```

```python
file.close()
```

```python
samples.shape
```

```python
num_rows = math.floor(math.sqrt(num_samples))
num_cols = num_samples // num_rows
num_cols += int(num_samples % num_rows > 0)
```

```python
figure =  make_subplots(
    rows=num_rows, 
    cols=num_cols,
    print_grid=True,
    shared_xaxes=True,
    shared_yaxes=True,
    horizontal_spacing=0.01,
    vertical_spacing=0.01,
    specs=[[{"type": "scene"}] * num_cols] * num_rows
)
```

```python
for i in range(num_rows):
    for j in range(num_cols):
        index = i * num_cols + j
        
        if index >= samples.shape[0]:
            continue
        figure.add_scatter3d(
            x=samples[index, :, 0],
            y=samples[index, :, 1],
            z=samples[index, :, 2], 
            marker=dict(
            cmin=seg_map.min(),
            cmax=seg_map.max(),
            color=seg_map[index],
            size=2
            ),
            mode="markers",
            row=i + 1,
            col=j + 1
        ) 
```

```python
figure.show()
```
